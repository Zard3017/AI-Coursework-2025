# 7.South African movies released in 2021

```python
south_africa_movies = df_netflix[(df_netflix['type'] == 'Movie') & (df_netflix['country'] == 'South Africa')]
south_africa_2021 = south_africa_movies[df_netflix['release_year'] == 2021]
print("South African movies released in 2021:")
print(south_africa_2021[['title', 'release_year']], "\n")
```

# 8.List the entire cast of all movies released in 2021.

```python
movies_2021 = df_netflix[(df_netflix['type'] == 'Movie') & (df_netflix['release_year'] == 2021)]

print("Cast of all movies released in 2021:\n")
for index, row in movies_2021.dropna(subset=['cast']).iterrows():
    print(f"{row['title']}: {row['cast']}")
```
