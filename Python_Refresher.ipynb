```python
#QUESTION 1
def days_seconds():
    days = int(input("Enter number of days: "))
    seconds = days * 24 * 60 * 60
    print(f"{days} days = {seconds} seconds")

# Call the function
days_seconds()
```

```python
#QUESTION 3
#function for area
def square_area(side):
    return side*side
#function for perimeter    
def square_perimeter(side):
    return 4*side 
length = float(input("Enter the side length of the square: "))
if length <=0:
    print("Please enter valid length")
else: 
    area= square_area(length)
    perimeter= square_perimeter(length)

print(f" the area of the square is :{area}")
print(f"the perimeter of the square is :{perimeter}")
```

```python
#QUESTION 4

def character_case(character):
    if character.isupper():
        print(f"The character {character} is uppercase")
    elif character.islower():
        print(f"The character {character} is lowercase")
    else:
        return f" Character is neither lowercase nor uppercase."

character_input = input("Enter a character: ")
result = character_case(character_input)
print(result)
              
```

```python
#QUESTION 6

numbers=[]
for i in range(5):
    value = float(input(f" Enter value {i+1} of 5:"))
    numbers.append(value)
average= sum(numbers)/len(numbers) 
print(f" The inputs are: {numbers}")
print(f"The average of the values is : {average}")
```
